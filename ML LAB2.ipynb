```python
import pandas as pd
df = pd.read_csv("student_data.csv")
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]
print("Students with Attendance > 90%:")
print(df[df["Attendance"] > 90])

print("\nStudents with TotalMarks > 90:")
print(df[df["TotalMarks"] > 90])

print("\nFemale Students:")
print(df[df["Gender"] == "F"])

print("\nMale Students:")
print(df[df["Gender"] == "M"])

print("\nStudents with ExternalMarks < 50:")
print(df[df["ExternalMarks"] < 50])
```

```python
import pandas as pd
df = pd.read_csv("student_data.csv")
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]

print("Student with Highest TotalMarks:")
print(df.loc[df["TotalMarks"].idxmax()])
print("\nStudent with Lowest TotalMarks:")
print(df.loc[df["TotalMarks"].idxmin()])

print("\nTop Three Performers:")
print(df.nlargest(3, "TotalMarks"))
print("\nBottom Three Performers:")
print(df.nsmallest(3, "TotalMarks"))

average_marks = df["TotalMarks"].mean()

print("\nClass Average TotalMarks:", average_marks)

count = len(df[df["TotalMarks"] > average_marks])

print("Number of students scoring above class average:", count)
```

```python
import pandas as pd
df = pd.read_csv("student_data.csv")
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]

male_avg = df[df["Gender"] == "M"]["TotalMarks"].mean()
print("Average TotalMarks of Male Students:", male_avg)
female_avg = df[df["Gender"] == "F"]["TotalMarks"].mean()
print("Average TotalMarks of Female Students:", female_avg)

attendance = df.groupby("Gender")["Attendance"].mean()

print("\nAverage Attendance of Male and Female Students:")
print(attendance)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file
df = pd.read_csv("student_data.csv")

# Create TotalMarks column
df["TotalMarks"] = df["InternalMarks"] + df["ExternalMarks"]

# 1. Bar Chart - TotalMarks of all students
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["TotalMarks"])
plt.title("Total Marks of Students")
plt.xlabel("Student Name")
plt.ylabel("Total Marks")
plt.show()

# 2. Histogram - Attendance
plt.figure(figsize=(8,5))
plt.hist(df["Attendance"], bins=5)
plt.title("Attendance Distribution")
plt.xlabel("Attendance")
plt.ylabel("Number of Students")
plt.show()

# 3. Pie Chart - Male and Female Students
gender_count = df["Gender"].value_counts()

plt.figure(figsize=(6,6))
plt.pie(gender_count, labels=gender_count.index, autopct="%1.1f%%", startangle=90)
plt.title("Distribution of Male and Female Students")
plt.show()

# 4. Scatter Plot - Attendance vs TotalMarks
plt.figure(figsize=(8,5))
plt.scatter(df["Attendance"], df["TotalMarks"])
plt.title("Attendance vs Total Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Total Marks")
plt.grid(True)
plt.show()
```
